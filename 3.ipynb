```python
import time

def display_instructions():
    print("Welcome to the Speed Typing Test!")
    print("You will be given a passage to type within a specified time limit.")
    print("Type the passage exactly as shown and press Enter to submit.")
    print("Get ready to start when you're prepared!\n")

def run_speed_typing_test():
    passage = "The quick brown fox jumps over the lazy dog."
    print("Type this passage within 10 seconds:\n")
    print(passage)
    
    input("Press Enter when you are ready to start.")
    
    start_time = time.time()
    user_input = input("Start typing: ")
    end_time = time.time()
    
    elapsed_time = end_time - start_time
    words_per_minute = len(passage.split()) / (elapsed_time / 60) if elapsed_time > 0 else 0
    
    if user_input == passage:
        print(f"Congratulations! You typed the passage correctly.")
        print(f"Your typing speed: {words_per_minute:.2f} words per minute")
    else:
        print("Sorry, you made a mistake or didn't finish in time.")
    
    accuracy = calculate_accuracy(passage, user_input)
    print(f"Accuracy: {accuracy:.2f}%")
    
    return words_per_minute, accuracy

def calculate_accuracy(original, typed):
    correct_chars = sum(1 for i in range(min(len(original), len(typed))) if original[i] == typed[i])
    total_chars = max(len(original), len(typed))
    accuracy = (correct_chars / total_chars) * 100 if total_chars > 0 else 0
    return accuracy

def main():
    display_instructions()
    
    num_rounds = int(input("Enter the number of rounds you want to play: "))
    print("\nLet's begin the test!\n")
    
    total_wpm = 0
    total_accuracy = 0
    
    for i in range(1, num_rounds + 1):
        print(f"Round {i}:")
        print("Get ready to start typing...")
        time.sleep(2)  # Pause for 2 seconds before starting each round
        wpm, accuracy = run_speed_typing_test()
        total_wpm += wpm
        total_accuracy += accuracy
    
    avg_wpm = total_wpm / num_rounds if num_rounds > 0 else 0
    avg_accuracy = total_accuracy / num_rounds if num_rounds > 0 else 0
    
    print("\nTest summary:")
    print(f"Average Words Per Minute: {avg_wpm:.2f}")
    print(f"Average Accuracy: {avg_accuracy:.2f}%")

if __name__ == "__main__":
    main()
```
